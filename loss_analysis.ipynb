```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_cache = pd.read_csv("./backtranslate_train_information_no_cache_clearing.csv")
df_no_cache = pd.read_csv("./backtranslate_train_information_with_cache_clearing.csv")
```

```python
plt.figure(figsize=(10, 6))

# Plotting train and validation losses for df_cache
plt.plot(df_cache['train_losses'], label='Train Loss (No Cache Clearing)', color='purple')
plt.plot(df_cache['validation_losses'], label='Validation Loss (No Cache Clearing)', color='purple', linestyle='dashed')

# Plotting train and validation losses for df_no_cache
plt.plot(df_no_cache['train_losses'], label='Train Loss (Cache Clearing)', color='green')
plt.plot(df_no_cache['validation_losses'], label='Validation Loss (Cache Clearing)', color='green', linestyle='dashed')

plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Train and Validation Losses Over Epochs')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))

# Plotting time per epoch for df_cache
plt.plot(df_cache['time_per_epoch'], label='Time per Epoch (No Cache Clearing)', color='purple')

# Plotting time per epoch for df_no_cache
plt.plot(df_no_cache['time_per_epoch'], label='Time per Epoch (Cache Clearing)', color='green')

plt.xlabel('Epoch')
plt.ylabel('Time per Epoch (seconds)')
plt.title('Time per Epoch Over Epochs')
plt.legend()
plt.grid(True)
plt.show()
```
